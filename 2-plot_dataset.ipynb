```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

font = {'family' : 'serif',
        #'weight' : 'bold',
         'size'   : 7.5
       }
matplotlib.rc('font', **font)
```

# Load the dataset using Pandas

```python
raw_df = pd.read_csv("dataset.csv")
```

```python
display(raw_df)
```

```python
df = raw_df.drop(labels=["x", "y", "length of tile size [in px]"], axis=1)
```

```python
display(df)
```

```python
print("The size of the dataset (Z) is", df.shape)
print("There are", df.shape[1], "columns")
print("There are", df.shape[0], "rows")
print("There are", df.shape[1] - 1, "features (n)")
print("There are", df.shape[0], "objects (N)")
```

# Extract the number of features

```python
print(np.array(df.columns))
```

```python
# The two lines below are equivalent
n = len(df.columns) - 1
n = df.shape[1] - 1

print("Number of features:", n)
```

# Extract the number of objects

```python
N = df.shape[0]
print("Number of features:", N)
```

# Extract the number of classes

```python
c = len(df["c"].unique())
print("Number of classes:", c)
```

# Extract the labels

```python
y = df["c"]
print(y)
```

# Extract the feature names

```python
# Get the label of all the columns, and remove the last one
dimensions = np.array(df.columns)[:-1]
print(dimensions)
```

# Display the scatterplot matrix of features

## Using Matplotlib

```python
plt.figure(figsize=(10, 10))

for j in range(n):
    for i in range(n):
        plot_index = j * n + i + 1
        plt.subplot(n, n, plot_index)
        plt.scatter(df[dimensions[i]], df[dimensions[j]], marker="x", c=y)
        
        if j == n - 1:
            plt.xlabel(dimensions[i])
            
        if i == 0:
            plt.ylabel(dimensions[j])
            
        plt.grid()
        
plt.savefig("scatterplot-matplotlib.png")
plt.savefig("scatterplot-matplotlib.pdf")
```

## Using Seaborn

```python
import seaborn as sns

plt.figure(figsize=(5, 17))
plot = sns.pairplot(df, hue="c")
plot.savefig("scatterplot-seaborn.png")
plot.savefig("scatterplot-seaborn.pdf")
```

## Using Pandas

```python
from pandas.plotting import scatter_matrix
```

```python
pd.plotting.scatter_matrix(df, c=y, figsize = [8,8],
                      s=50, marker = '.');

plt.savefig("scatterplot-pandas.png")
plt.savefig("scatterplot-pandas.pdf")
```
